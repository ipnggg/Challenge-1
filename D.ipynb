**D. Does ordered egg whites instead of egg increase a meal's nutritional value**

```python
df_egg = df[df['Item'].str.contains('Egg')]
df_egg=df_egg[['Item','Calories','Total Fat','Carbohydrates','Protein']]
```

```python
df_egg.reset_index(inplace=True,drop=True)
df_egg
```

```python
df_egg = df_egg.iloc[[2,3,9,11,14,15,18,19],] #bebas pilih sampel yang mana aja
df_egg
```

```python
df_egg.set_index('Item', drop= True, inplace=True)
df_egg
```

```python
df_egg = df_egg / df_egg.max()
df_egg
```

```python
ax = df_egg.plot(kind='bar',figsize=(20,10))
ax.set_title('Nutritional Value Comparison')
ax.set_ylabel('Value-Normalized')
ax.set_xlabel('Item')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show
```
