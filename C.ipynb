**C. Does ordered grilled chicken instead of crispy increase a sandwich's nutritional value**

```python
df_sandwich = df[df['Item'].str.contains('Sandwich')]
```

```python
df_sandwich=df_sandwich[['Item','Calories','Total Fat','Carbohydrates','Protein']]
df_sandwich.index = df_sandwich['Item']
```

```python
df_sandwich.drop(['Item'], axis = 1, inplace=True)
df_sandwich.drop(['Southern Style Crispy Chicken Sandwich'], axis = 0, inplace=True)
```

```python
df_sandwich = df_sandwich / df_sandwich.max()
```

```python
ax = df_sandwich.plot(kind='bar',figsize=(20,10))
ax.set_title('Nutritional Value Comparison')
ax.set_ylabel('Value-Normalized')
ax.set_xlabel('Item')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show
```
